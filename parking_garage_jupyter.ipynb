```python
from IPython.display import clear_output as clear

class Garage():

    def __init__(self, tickets, parkingSpaces):
        self.tickets = list(range(1, tickets + 1)) # tickets available, correlates with parkingSpaces.
        self.parkingSpaces = list(range(1, parkingSpaces + 1)) # paid/unpaid
        self.currentTicket = {} # keys: tickets,
    

    def takeTicket(self): # take a ticket and add to currentTicket, take out from tickets and parkingSpaces
        clear()
        if not self.tickets:
            print("⛔️ This parking garage is full! ⛔️") 
        else:
            tix = self.tickets[0]
            print(f"Your parking spot and ticket number is #{tix}. 🎟")
            self.tickets.remove(tix)
            self.currentTicket[tix] = "Not Paid"
            print(f"\nParking spots taken: {self.currentTicket}")
            self.parkingSpaces.remove(tix)
            print(f"\nParking spaces remaining: {self.parkingSpaces}")





    def payForParking(self):  # check if ticket was paid, provide option to pay
        clear()
        print(f"\nHere's the list of tickets: {self.currentTicket}")
        select_ticket = int(input("\nEnter your parking spot number 🎟: ")) 
        if self.currentTicket[select_ticket] == "Not Paid":
            print(f"\nYour ticket: #{select_ticket} has not been paid for 🛑. ")
            pay_tix = input("Type 'pay' to pay for your Ticket 💵. ")
            if pay_tix.lower() == 'pay':
                clear()
                print(f"\nThank you! ticket #{select_ticket} is now paid for 🤑!")
                self.currentTicket[select_ticket] = 'Paid'
            else: 
                print("\nYou have not paid for your ticket: please try again: " )
        elif self.currentTicket[select_ticket] == 'Paid':
            print("This ticket has already been paid for 🤑")
        else:
            print(f"\nThis parking spot is not in use at this time 🧐.. \nPlease enter '2' to try again: ")


    def leaveGarage(self):  # check if ticket is paid and leave garage
        clear()
        print(self.currentTicket)
        check_tix = int(input("What is your ticket number? 🎟 "))
        if check_tix not in self.currentTicket: 
            clear()
            print("🛑 That is not an active ticket: please note list below and try again: ")
            print(self.currentTicket)
        elif self.currentTicket[check_tix] == 'Not Paid':
            print("You need to pay for your ticket 💵")
        elif self.currentTicket[check_tix] == 'Paid':
            clear()
            print("\nYour Ticket is paid 🤑 Thank You, have a nice day!")
            self.tickets.append(check_tix)
            self.tickets.sort()
            self.parkingSpaces.append(check_tix)
            self.parkingSpaces.sort()
            del self.currentTicket[check_tix]
            print(f"{self.parkingSpaces} parking spots available at this time.")


    def run_garage(self):
        clear()
        while True:
            print("""

            Welcome to the parking garage. 🚗 🚙
            What would you like to do?
            [1] park 🎟
            [2] pay 💵
            [3] leave 🚙💨
            
            """)

            response = int(input("Choose your option: "))
            if response == 1 :
                self.takeTicket()
            elif response == 2 :
                self.payForParking()
            elif response == 3 :
                self.leaveGarage()

            else: 
                print("That is not a valid option: please choose 1, 2, or 3")


TheShoppesGarage = Garage(15,15)

TheShoppesGarage.run_garage()
```
